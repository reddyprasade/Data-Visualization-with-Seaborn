#### Statistical estimation and error bars

* Data visualization sometimes involves a step of aggregation or estimation, where multiple data points are reduced to a summary statistic such as the `mean` or `median`.
* When showing a `summary statistic`, it is usually appropriate to add `error bars`, which provide a visual cue about how well the summary represents the underlying data points.

![](https://seaborn.pydata.org/_images/error_bars_2_0.svg)

`Errorbar`: The error bars around an estimate of `central tendency` can show one of two general things
* `Spread:` Given the same sample size, estimates will be more uncertain when data has a broader spread.
* `uncertainty:` But uncertainty will decrease as sample sizes grow, whereas spread will not.

* `Parametric:`One approach is parametric, using a formula that relies on assumptions about the `shape of the distribution`
* `nonparametric:` The other approach is nonparametric, using only the data that you provide.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def plot_errorbars(arg, **kws):
    np.random.seed(sum(map(ord, "error_bars")))
    x = np.random.normal(0, 1, 100)
    f, axs = plt.subplots(2, figsize=(7, 2), sharex=True, layout="tight")
    sns.pointplot(x=x, errorbar=arg, **kws, capsize=.3, ax=axs[0])
    sns.stripplot(x=x, jitter=.3, ax=axs[1])
```

```python
plot_errorbars('sd')
```

```python
plot_errorbars(("pi", 50))
```

```python
plot_errorbars("se")
```

```python
plot_errorbars("ci")
```

```python
plot_errorbars(("se", 2))
```

```python
plot_errorbars("ci", estimator="median")
```

```python
plot_errorbars("ci", n_boot=5000, seed=10)
```

```python
plot_errorbars(lambda x: (x.min(), x.max()))
```
