```python
# Import seaborn
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
# Apply the default theme
sns.set_theme()
```

```python
# Load an example dataset
import pandas as pd
tips = pd.read_csv("Datasets/tips.csv")
tips.head()
```

```python
tips.info()
```

```python
tips.time.value_counts()
```

```python
sns.regplot(data=tips,x='total_bill',y='tip')
```

```python
sns.relplot(data=tips,
            x='total_bill',
            y='tip',
            hue='smoker',
            style="smoker",
            size="size",
            col="time")
```

```python
sns.relplot(data=tips,
            x='total_bill',
            y='tip',
            #hue='smoker',
            #style="smoker",
            #size="size",
            kind="line",)
            #col="time")
```

Statistical estimation in seaborn goes beyond descriptive statistics. For example, it is possible to enhance a scatterplot by including a linear regression model (and its uncertainty) using lmplot():

```python
sns.lmplot(data=tips, x="total_bill", y="tip", col="time", hue="smoker")
```

### Distributional representations

![](https://miro.medium.com/v2/resize:fit:962/1*DmPUIjvecL7KllOamoFSDw.png)

```python
sns.displot(data=tips,x="total_bill",col='time',kde=True)
```

```python
# Seaborn also tries to promote techniques that are powerful but less familiar, such as calculating and plotting the empirical cumulative distribution function of the data:
sns.displot(data=tips,x="total_bill",col='time',kind='ecdf',hue='smoker',rug=True)
```

### Plots for categorical data

```python
sns.catplot(data=tips,kind='swarm',x='day',y='total_bill',hue='smoker')
```

```python
sns.catplot(data=tips,kind='violin',x='day',y='total_bill',hue='smoker')
```

```python
sns.catplot(data=tips,kind='bar',x='day',y='total_bill',hue='smoker')
```

```python
sns.jointplot(data=tips,x="total_bill",y='tip',hue="smoker")
```

```python
sns.pairplot(data=tips,hue = 'smoker')
```
